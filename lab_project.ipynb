```python
import numpy as np

def encrypt_text(text, key):
    ##Encrypt a string using XOR and return a  1D numpy array.   
    ascii_values = np.array([ord(c) for c in text], dtype=np.uint16)
    encrypted = ascii_values ^ np.array(key)
    return encrypted

def encrypt_file_to_txt(input_file, output_file, key):
    """
    Encrypt each line of a text file into a 1D encrypted array.
    Each encrypted 1D array is stored as comma-separated values.
    """
    with open(input_file, "r", encoding="utf-8") as f:
        lines = f.readlines()

    with open(output_file, "w", encoding="utf-8") as out:
        for line in lines:
            line = line.rstrip("\n")

            if not line:
                out.write("\n")
                continue

            encrypted = encrypt_text(line, key)
            out.write(",".join(map(str, encrypted)) + "\n")

    print("Encryption completed →", output_file)



def decrypt_array(arr, key):  
    ##Decrypt a 1D numpy array and return the original string.  
    decrypted = arr ^ np.array(key)
    p=''.join(chr(int(v)) for v in decrypted)
    return p


def decrypt_file_from_txt(input_file, output_file, key):
    
    ##Reads encrypted 1D arrays from TXT and writes decrypted text lines.
    
    decrypted_lines = []

    with open(input_file, "r", encoding="utf-8") as f:
        lines = f.readlines()

    for line in lines:
        line = line.strip()

        if not line:
            decrypted_lines.append("")
            continue

        numbers = np.array(list(map(int, line.split(","))))
        text_line = decrypt_array(numbers, key)
        decrypted_lines.append(text_line)

    with open(output_file, "w", encoding="utf-8") as f:
        f.write("\n".join(decrypted_lines))

    print("1D Decryption completed →", output_file)

```
